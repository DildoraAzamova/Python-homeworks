```python

# Create custom modules
# math_operations.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b


# string_utils.py

def reverse_string(s):
    return s[::-1]

def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)


# Create Custom Packages

import math

def calculate_area(radius):
    return math.pi * radius * radius

def calculate_circumference(radius):
    return 2 * math.pi * radius

# file_reader.py
def read_file(file_path):
    try:
        with open(file_path, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return "File not found."
    
# file_writer.py
def write_file(file_path, content):
    with open(file_path, 'w') as f:
        f.write(content)


# Example Main Program to Use All

from math_operations import add, divide
from string_utils import reverse_string, count_vowels
from geometry.circle import calculate_area, calculate_circumference
from file_operations.file_writer import write_file
from file_operations.file_reader import read_file

# Math
print(add(10, 5))
print(divide(10, 2))

# String
print(reverse_string("hello"))
print(count_vowels("hello world"))

# Geometry
print(calculate_area(3))
print(calculate_circumference(3))

# File Operations
write_file("test.txt", "Hello from file operations!")
print(read_file("test.txt"))
```
