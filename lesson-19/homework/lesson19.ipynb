```python
# Homework 1. Task 1

result=df.groupby('Category').agg(
    TotalQuantitySold=('Quantity', 'sum'),
    AveragePricePerUnit=('Price', 'mean'),
    MaxQuantitySold=('Quantity', 'max')
)

result
```

```python
# Homework 1. Task 2

# Group by Category and Product, then sum Quantity
groupped=df.groupby(['Category','Product'])['Quantity'].sum().reset_index()

# For each Category, find the Product with the max total Quantity
TopSelling=groupped.sort_values('Quantity', ascending=False).groupby('Category').first().reset_index()

TopSelling
```

```python
# Homework 2. 

import pandas as pd

df1 = pd.read_csv(r'C:\Users\HP\Desktop\Python\Data\customer_orders.csv')
df1
```

```python
# Homework 2. Task 1

# Step 1: Count orders per customer
order_counts=df1.groupby('CustomerID').size().reset_index(name='OrderCounts')

# Step 2: Filter customers with 20 or more orders
active_customers=order_counts[order_counts['OrderCounts']>=20]

# Step 3: Filter original DataFrame to only include those customers
filterd_df=df1[df1['CustomerID'].isin(active_customers['CustomerID'])]

filterd_df
```

```python
# Homework 2. Task 2

# Calculate unit price
df1['UnitPrice']=df1['Price']/df1['Quantity']

# Group by CustomerID and calculate average unit price
avg_price_per_customer=df1.groupby('CustomerID')['UnitPrice'].mean().reset_index()

# Filter customers with avg unit price > $120
high_value_customers=avg_price_per_customer[avg_price_per_customer['UnitPrice']>120]

high_value_customers
```

```python
# Homework 2. Task 3

# Group by product and calculate total quantity and total price
product_summary=df1.groupby('OrderID').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalPrice=('Price', 'sum')
).reset_index()

# Filter out products with total quantity < 5
filtered_product=product_summary[product_summary['TotalQuantity']<=5]

filtered_product
```

```python
# Homework 3

import pandas as pd
import sqlite3 

# STEP 1: Connect to SQLite database and read data
conn=sqlite3.connect(r'C:\Users\HP\Desktop\Python\Data\population.db')

# Load population table using SQL
population_df = pd.read_sql_query("SELECT * FROM population", conn)

conn.close()  # good practice to close connection

# STEP 2: Load salary band definitions from Excel
salary_band_df = pd.read_excel(r'C:\Users\HP\Desktop\Python\Data\population_salary_analysis.xlsx')

# Preview your dataframes
population_df
salary_band_df 
```

```python
# STEP 3: Define salary bands

# Create a function to assign salary category
def categorize_salary(salary, bands):
    for _, row in bands.iterrows():
        if row['Min']<=salary<=row['Max']:
            return row['Category']
    return 'Unknown'

# Apply the salary band category
population_df['Salary_Category'] = population_df['salary'].apply(lambda x: categorize_salary(x, salary_band_df))
```

```python
# STEP 4: Calculate Overall Measures

# Total population
total_population=len(population_df)

# Group by salary category
grouped=population_df.groupby('Salary_Category')

# Calculate measures
summary=grouped['Salary'].agg(
    Count='count',
    Average='mean',
    Median='Median'
).reset_index()

summary['Percentage']=(summary['Count']/total_population*100).round(2)

summary
```

```python
# STEP 5: Measures Per State & Salary Category

# Group by State and Salary Category
state_grouped = population_df.groupby(['state', 'Salary_Category'])

# Calculate per-state metrics
state_summary = state_grouped['Salary'].agg(
    Count='count',
    Average='mean',
    Median='median'
).reset_index()

# Add percentage column per state
state_totals = population_df.groupby('State').size().rename("Total_Pop")
state_summary = state_summary.merge(state_totals, on='State')
state_summary['Percentage'] = (state_summary['Count'] / state_summary['Total_Pop'] * 100).round(2)

print("ðŸ“Š Per-State Salary Category Analysis:")
print(state_summary.drop(columns='Total_Pop'))

```
